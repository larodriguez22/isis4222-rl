## Ambiente 

Para la definici贸n del agente manejaremos una cuadricula de `nxn` como nuestro ambiente. El ambiente tiene obstaculos, es decir casillas por las cuales no se puede pasar, una casilla de inicio, y algunas casillas de terminaci贸n. Este ambiente, es conocido en el aprendizaje por refuerzo como el ambiente de Gridworld. Un ejemplo del ambiente para el caso `5x5` se muestra a continuaci贸n.

![gridworld.png](./img/gridworld.png)

Een este ejemplo del ambiente el agente comienza en la casilla inferior izquierda y tiene como objetivo llegar a la casilla de salida verde, con recompensa 1. La otra casilla de salida, tiene recompensa -1

**Task 1.** Complete la definici贸n del ambiente estableciendo por:
1. La cuadr铆cula (`board`), con las dimensiones dadas por par谩metro. Definiremos las casillas por las que puede pasar el agente como espacios en blanco y las casillas por las que no puede pasar el agente con un `'*'`.
2. Un atributo (`dimensions`) para almacenar las dimensiones del tablero
3. Las recompensas de cada casilla de la cuadr铆cula estan definidas dentro de la definici贸n de `board`.
    - +1 para la casilla objetivo 
    - -1 para las casillas de trampa
4. Un atributo con el estado actual (`current_state`) en el que se encuentra el agente. Por defecto este estado ser谩 la posici贸n (0,0).

#### Comportamiento del ambeinte

Una vez definido el ambiente definimos su comportamiento. Para ello defina los m茅todos
1. `get_current_state` que retorna el estado actual (la casilla donde se encuentra el agente)
2. `get_posible_actions` que retorna las acciones disponibles para cada estado, dado por par谩metro como una tupla `(i,j)`. Las acciones estar谩n dadas por su nombre (`'up', 'down', 'left', 'right'`)
3. `do_action` que recibe como par谩metro la acci贸n a ejecutar y retorna el valor de la recompensa y el nuevo estado del agente, como un pareja `(reward, new_state)`
4. `reset` que restablece el ambiente a su estado inicial (en nuestro caso moviendo al agente a la posici贸n (0, 0))
5. `is_terminal` que termina si est谩 en el estado final o no. en nuestro caso el estado final estar谩 determinado por la casilla con recompensa igual a 1.



```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

class Gridworld:
    def __init__(self, board, dimensions, initial_state = (0,0)):
        self.shape = dimensions
        self.nrows, self.ncols = dimensions
        self.grid = np.zeros(dimensions)
        
        for i in range(len(board)):
            for j in range(len(board[0])):
                if board[i][j] == "*":
                    self.grid[i][j] = -2
                elif board[i][j] == "+1":
                    self.grid[i][j] = 1
                elif board[i][j] == "-1":
                    self.grid[i][j] = -1
        # self.grid[-1, -1] = 1  # set the final state to have a reward of 1
        # print("grid despues",self.grid)
        self.initial_state = initial_state
        self.state = initial_state

    def get_current_state(self):
        return self.state

    def get_possible_actions(self, state):
        i, j = state
        actions = []
        if i > 0 and self.grid[i-1][j] != -2:
            actions.append('up')
        if i < self.nrows - 1 and self.grid[i+1][j] != -2:
            actions.append('down')
        if j > 0 and self.grid[i][j-1] != -2:
            actions.append('left')
        if j < self.ncols - 1 and self.grid[i][j+1] != -2:
            actions.append('right')
        return actions

    def do_action(self, action):
        i, j = self.state
        if action == 'up' :
            i -= 1
        elif action == 'down':
            i += 1
        elif action == 'left':
            j -= 1
        elif action == 'right':
            j += 1
        self.state = (i, j)
        return self.grid[i, j], self.state

    def reset(self):
        self.state = self.initial_state

    def is_terminal(self):
        return self.grid[self.state] == 1
    
    def plot(self):
        fig1 = plt.figure(figsize=(10, 10))
        ax1 = fig1.add_subplot(111, aspect='equal')
        
        # Lineas
        for i in range(0, len(self.grid)+1):
            ax1.axhline(i , linewidth=2, color="#2D2D33")
            ax1.axvline(i , linewidth=2, color="#2D2D33")
        
        # Amarillo - inicio
        ax1.add_patch(patches.Rectangle(self.initial_state, 1, 1, facecolor = "#F6D924"))
        for i in range(len(self.grid)):
            for j in range(len(self.grid[0])):
                if self.grid[i][j] == 1: # verde
                    ax1.add_patch(patches.Rectangle((i,j), 1, 1, facecolor = "#68FF33"))
                if self.grid[i][j] == -2: # gris
                    ax1.add_patch(patches.Rectangle((i,j), 1, 1, facecolor = "#6c7780"))
                if self.grid[i][j] == -1: # rojo
                    ax1.add_patch(patches.Rectangle((i,j), 1, 1, facecolor = "#cc0000"))
        plt.scatter(self.state[0] + 0.5, self.state[1] +0.5, s = 100, color = "black", marker = "o", facecolor = "blue", edgecolors = "blue", zorder = 10)
        for i in range(len(self.grid)):
            for j in range(len(self.grid[0])):
                if self.grid[i, j] == -2:
                    ax1.text(i+0.5, j+0.5, "*", ha='center', va='center')
                else:
                    ax1.text(i+0.5, j+0.5, self.grid[i, j], ha='center', va='center')
        plt.axis("off")
        plt.show()
        
    
```

```python
board = [  [' ', '-1', ' ', '*', ' '],
  [' ', '*', ' ', ' ', ' '],
  [' ', ' ', ' ', ' ', ' '],
  [' ', ' ', ' ', ' ', ' '],
  [' ', ' ', ' ', ' ', '+1']
]

dimensions = (5, 5)


env = Gridworld(board, dimensions)

# get the current state
print(env.get_current_state()) # (0, 0)

# get the possible actions
print(env.get_possible_actions(env.get_current_state())) # ['right', 'down']

# do an action
print(env.do_action('right')) # (0, (0, 1))

# get the current state after doing an action
print(env.get_current_state()) # (0, 1)

# check if the environment is in a terminal state
print(env.is_terminal()) # False

# reset the environment to its initial state
env.reset()
print(env.get_current_state()) # (0, 0)

env.plot()
```

```python
import numpy as np

def expected_reward(state, grid, rewards, transition_prob=0.25):
    i, j = state
    possible_actions = ['up', 'down', 'left', 'right']
    nrows, ncols = len(grid), len(grid[0])
    expected_reward = 0
    print("--------------")
    print("state", state)
    for action in possible_actions:
        if action == 'up': 
            if i > 0 and grid[i-1][j] != '*':
                expected_reward += transition_prob * rewards[i-1][j]
                print("up", expected_reward)
                print("Estado futuro", grid[i-1][j])
        elif action == 'down':
            if i < nrows - 1 and grid[i+1][j] != '*':
                expected_reward += transition_prob * rewards[i+1][j]
                print("down", expected_reward)
                print("Estado futuro", grid[i+1][j])
        elif action == 'left':
            if j > 0 and grid[i][j-1] != '*':
                expected_reward += transition_prob * rewards[i][j-1]
                print("left", expected_reward)
                print("Estado futuro", grid[i][j-1])
        elif action == 'right':
            if j < ncols - 1 and grid[i][j+1] != '*':
                expected_reward += transition_prob * rewards[i][j+1]
                print("right", expected_reward)
                print("Estado futuro", grid[i][j+1])
    return expected_reward

grid = [
    [' ', ' ', ' ', ' ', ' ',' ', ' ', ' ', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', '*', '*', '*','*', '*', '*', ' ', ' '],
    [' ', ' ', '-1', ' ', '+1','-1', ' ', ' ', ' ', ' '],
    [' ', ' ', '-1', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', ' ', ' ', ' '],
]

rewards = np.zeros((10, 10))
rewards[5][4] = 1

for i in range(10):
    for j in range(10):
        if grid[i][j] != '*':
            rewards[i][j] = -1

expected_rewards = np.zeros((10, 10))
for i in range(10):
    for j in range(10):
        if grid[i][j] != '*':
            expected_rewards[i][j] = expected_reward((i, j), grid, rewards)

print()
print("Tabla de recompensas esperadas")
print(expected_rewards)
```
