```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
import time

class Gridworld:
    def __init__(self, board, initial_state = (0,0)):
        self.nrows, self.ncols = len(board),len(board[0])
        self.grid = np.zeros((self.nrows, self.ncols))
        self.is_terminal = False
        
        for i in range(len(board)):
            for j in range(len(board[0])):
                if board[i][j] == "*":
                    self.grid[i][j] = -2
                elif board[i][j] == "+1":
                    self.grid[i][j] = 1
                elif board[i][j] == "+100":
                    self.grid[i][j] = 100
                elif board[i][j] == "-1":
                    self.grid[i][j] = -1
                elif board[i][j] == "-100":
                    self.grid[i][j] = -100
        # self.grid[-1, -1] = 1  # set the final state to have a reward of 1
        # print("grid despues",self.grid)
        self.initial_state = initial_state
        self.state = initial_state
        
    def get_states(self):
        return [(i, j) for i in range(len(self.grid)) for j in range(len(self.grid[0]))]

    def get_current_state(self):
        return self.state

    def get_possible_actions(self, state):
        i, j = state
        actions = []
        if self.grid[state]==0 or self.grid[state]==-1:
            if j > 0 and self.grid[i][j-1] != -2:
                actions.append('down')
            if j < self.ncols - 1 and self.grid[i][j+1] != -2:
                actions.append('up')
            if i > 0 and self.grid[i-1][j] != -2:
                actions.append('left')
            if i < self.nrows - 1 and self.grid[i+1][j] != -2:
                actions.append('right')
        elif self.grid[state] != -2:
            actions.append("end")
        elif self.grid[state] == -2:
            return [None]
        return actions
    
    def get_next_states(self, state, action):
        i, j = state
        reward = 0
        if action == 'left' :
            i -= 1
        elif action == 'right':
            i += 1
        elif action == 'down':
            j -= 1
        elif action == 'up':
            j += 1
        elif action == 'end':
            reward = self.grid[i, j]
        self.state = (i, j)
        return reward, self.state

    def do_action(self, action):
        i, j = self.state
        reward = 0
        done = False
        if action == 'left' :
            i -= 1
        elif action == 'right':
            i += 1
        elif action == 'down':
            j -= 1
        elif action == 'up':
            j += 1
        elif action == 'end':
            reward = self.grid[i, j]
            done = True
            
        self.state = (i, j)
        return self.grid[i, j], self.state, done

    def reset(self):
        self.state = self.initial_state

    def is_terminal(self):
        return self.grid[self.state] == 1
    
    def plot(self):
        fig1 = plt.figure(figsize=(10, 10))
        ax1 = fig1.add_subplot(111, aspect='equal')
        
        # Lineas
        for i in range(0, len(self.grid)+1):
            ax1.axvline(i , linewidth=2, color="#2D2D33")
        for j in range(0, len(self.grid[0])+1):
            ax1.axhline(j , linewidth=2, color="#2D2D33")
        
        # Amarillo - inicio
        ax1.add_patch(patches.Rectangle(self.initial_state, 1, 1, facecolor = "#F6D924"))
        for i in range(len(self.grid)):
            for j in range(len(self.grid[0])):
                if self.grid[i][j] >= 1: # verde
                    ax1.add_patch(patches.Rectangle((i,j), 1, 1, facecolor = "#68FF33"))
                if self.grid[i][j] == -2: # gris
                    ax1.add_patch(patches.Rectangle((i,j), 1, 1, facecolor = "#6c7780"))
                if self.grid[i][j] == -1 or self.grid[i][j] == -100: # rojo
                    ax1.add_patch(patches.Rectangle((i,j), 1, 1, facecolor = "#cc0000"))
        plt.scatter(self.state[0] + 0.5, self.state[1] +0.5, s = 100, color = "black", marker = "o", facecolor = "blue", edgecolors = "blue", zorder = 10)
        for i in range(len(self.grid)):
            for j in range(len(self.grid[0])):
                if self.grid[i, j] == -2:
                    ax1.text(i+0.5, j+0.5, "*", ha='center', va='center')
                else:
                    ax1.text(i+0.5, j+0.5, self.grid[i, j], ha='center', va='center')
        plt.axis("off")
        plt.show()
        
    def get_action_test(self, action):
        if action=='left':
            return '<'
        if action=='right':
            return '>'
        if action=='up':
            return '^'
        if action=='down':
            return 'v'
        if action=='exit':
            return 'x'
        
    def plot_action(self, actions, values):
        fig1 = plt.figure(figsize=(10, 10))
        ax1 = fig1.add_subplot(111, aspect='equal')
        
        # Lineas
        for i in range(0, len(self.grid)+1):
            ax1.axvline(i , linewidth=2, color="#2D2D33")
        for j in range(0, len(self.grid[0])+1):
            ax1.axhline(j , linewidth=2, color="#2D2D33")
        
        # Amarillo - inicio
        ax1.add_patch(patches.Rectangle(self.initial_state, 1, 1, facecolor = "#F6D924"))
        for i in range(len(self.grid)):
            for j in range(len(self.grid[0])):
                if self.grid[i][j] >= 1: # verde
                    ax1.add_patch(patches.Rectangle((i,j), 1, 1, facecolor = "#68FF33"))
                if self.grid[i][j] == -2: # gris
                    ax1.add_patch(patches.Rectangle((i,j), 1, 1, facecolor = "#6c7780"))
                if self.grid[i][j] == -1 or self.grid[i][j] == -100: # rojo
                    ax1.add_patch(patches.Rectangle((i,j), 1, 1, facecolor = "#cc0000"))
        self.reset()
        plt.scatter(self.state[0] + 0.5, self.state[1] +0.5, s = 100, color = "black", marker = "o", facecolor = "blue", edgecolors = "blue", zorder = 10)
        
        for i in range(len(self.grid)):
            for j in range(len(self.grid[0])):
                if self.grid[i, j] == -2:
                    ax1.text(i+0.5, j+0.5, "*", ha='center', va='center')
                else:
                    try:
                        ax1.text(i+0.5, j+0.75, round(values[i, j],2), ha='center', va='center')
                        text2=self.get_action_test(actions[(i,j)])
                        ax1.text(i+0.5, j+0.25, text2, ha='center', va='center')
                    except:
                        ax1.text(i+0.5, j+0.75, 0, ha='center', va='center')
                        ax1.text(i+0.5, j+0.25, "None", ha='center', va='center')
        plt.axis("off")
        plt.show()
        
```

# SARSA: State-Action-Reward-State-Action

En este ejercicio vamos a implementar el algoritmo de SARSA como un ejemplo de los métodos on-policy para el aprendizaje por refuerzo. Esto es, crear agentes que aprenden a alcanzar un objetivo específico.

El método de SARSA se basa en el cálculo de los q-valores utilizando los valores calculados para el estado de llegada siguiendo la fórmula de actualización de los q-valores:

$Q(s,a) \leftarrow (1-\alpha)Q(s,a) + \alpha[R(s) + \gamma Q(s',a')] $

Para implementar SARSA definiremos un agente, `sarsa_agent.py` el cual utilizaremos para interactuar con el ambiente de Gridworld.

**Task 1**
1.	Implemente la classe `SARSA` con cinco atributos:
 - `epsilon` que corresponde a la estrategia de aprendizaje $\epsilon$-greedy, `0.9` por defecto.
 -`gamma` que corresponde al factor de decuento a utilizar, `0.96` por defecto.
 -`alpha` que corresponde a la taza de aprendizajem `0.81` por defecto.
 -`Q` que almacena los q-valores del agente.
 -`env` que es una referencia al ambiente.

2. El comportamiento del agente (la interacción con el ambiente) esta dado por los métodos:
    - `choose_action` que recibe un estado como parámetro y retorna la acción a ejecutar para dicho estado siguiendo una estrategia $\epsilon$-greedy.
    - `action_function` que recibe como parámetro los componentes de SARSA (estado1, acción1, recompensa, estado2, acción2) y calcula el q-valor `Q(estdo1, acción1)`.

3. La interacción entre el agente y el ambiente inicia desde el ambiente, que ejecuta cada interacción de SARSA para cada episodio.
(1) La interacción comienza decidiendo la acción a tomar para el estado actual (la cual esta dada por el agente), (2) luego debemos ejecutar la acción, obteniendo el estado de llegada y la recompensa de ejecutar dicha acción, (3) luego calculamos la acción a tomar para el estado de llegada, (4) por último calculamos el q-valor definido por la función de las acciones.

```python
class SARSA:
    def __init__(self, env, epsilon=0.9, gamma=0.96, alpha=0.81):
        # Initialize the SARSA algorithm with hyperparameters
        self.epsilon = epsilon # Exploration rate
        self.gamma = gamma # Discount factor
        self.alpha = alpha # Learning rate
        self.env = env # Environment to interact with
        self.Q = {} # Q table to store state-action values

        # Initialize Q table with 0 for each state-action pair
        for state in self.env.get_states():
            for action in self.env.get_possible_actions(state):
                self.Q[(state, action)] = 0
        
    def choose_action(self, state):
        # Epsilon-greedy policy: choose random action with probability epsilon, otherwise choose action with max Q value
        if np.random.rand() < self.epsilon:
            actions = self.env.get_possible_actions(state)
            action = actions[np.random.randint(len(actions))]
        else:
            values = np.array([self.Q[(state, action)] for action in self.env.get_possible_actions(state)])
            max_indices = np.where(values == np.max(values))[0]
            # print(max_indices)
            action = self.env.get_possible_actions(state)[np.random.choice(max_indices)]
        
        return action
    
    def action_function(self, state1, action1, reward, state2, action2):
        # Update Q value for state-action pair based on TD error
        q_current = self.Q[state1, action1] # Current Q value
        q_next = self.Q[state2, action2] # Q value of next state-action pair
        td_target = reward + self.gamma * q_next # TD target
        td_error = td_target - q_current # TD error
        new_q = q_current + self.alpha * td_error # New Q value
        self.Q[state1, action1] = new_q # Update Q table
        
    def run(self,num_episodes):
        # Train the SARSA algorithm for a certain number of episodes
        for i in range(num_episodes):
            self.env.reset() # Reset environment
            state = self.env.get_current_state() # Get current state
            action = self.choose_action(state) # Choose initial action
            done = False
            while not done:
                reward, next_state, done= self.env.do_action(action) # Take action and observe next state and reward
                next_action = self.choose_action(next_state) # Choose next action
                self.action_function(state, action, reward, next_state, next_action) # Update Q table
                state = next_state # Update current state
                action = next_action # Update current action
        return self.Q
    
    def best_action(self, state):
        # Choose action with highest Q value for a given state
        values = np.array([self.Q[(state, action)] for action in self.env.get_possible_actions(state)])
        max_indices = np.where(values == np.max(values))[0]
        action = self.env.get_possible_actions(state)[max_indices[0]]
        return action
    
    def test_performance(self):
        # Evaluate the performance of the trained SARSA algorithm on the environment
        done = False
        actions = {}
        self.env.reset()
        values = np.zeros_like(self.env.grid)
        for state in self.env.get_states():
        # while not done:
        #     state = self.env.get_current_state()
            action = self.best_action(state)
            actions[state] = action
            values[state] = self.Q[state, action]
            # state, reward, done = self.env.do_action(action)
        print(actions)
        self.env.plot_action(actions, values)
```

```python
# create the Gridworld environment
board = [
    [' ', ' ', ' ', ' ', ' ',' ', ' ', ' ', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', '*', '*', '*','*', '*', '*', ' ', ' '],
    [' ', ' ', '-1', ' ', '+1','-1', ' ', ' ', ' ', ' '],
    [' ', ' ', '-1', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', '*', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ',' ', ' ', ' ', ' ', ' '],
]

initial_state = (0, 9)

gridworld = Gridworld(board, initial_state)

# gridworld.plot()
```

Si se utiliza un factor de descuento de 1 (γ=1) en el algoritmo SARSA de aprendizaje por refuerzo, esto significa que el agente no considera el futuro en su toma de decisiones y solo se centra en las recompensas inmediatas. Sin embargo, aqui se necesita el descuento, pues es necesario para garantizar que el agente aprenda a maximizar las recompensas a largo plazo. Si el descuento es demasiado bajo, el agente puede caer en un comportamiento miope y subóptimo, y no logrará maximizar la recompensa total.

Para minimizar la trayectoria del agente entre el estado inicial y el estado de llegada en el algoritmo SARSA, podemos tomar las siguientes medidas:
- Elegir una política adecuada: La política es la estrategia que el agente utiliza para elegir sus acciones. Una buena política puede ayudar al agente a avanzar en la dirección correcta y evitar acciones que lo alejen del estado de llegada. Por lo tanto, elegir una política adecuada es importante para minimizar la trayectoria del agente.

- Ajustar el factor de aprendizaje: El factor de aprendizaje es un parámetro que controla la rapidez con la que el agente aprende. Si el factor de aprendizaje es demasiado alto, el agente puede oscilar entre acciones subóptimas y no aprender eficazmente. Si el factor de aprendizaje es demasiado bajo, el agente puede tardar demasiado en aprender. Por lo tanto, ajustar el factor de aprendizaje de manera adecuada puede ayudar a minimizar la trayectoria del agente.

- Elegir un factor de descuento adecuado: El factor de descuento temporal γ determina la importancia que el agente da a las recompensas futuras. Si el factor de descuento temporal es demasiado bajo, el agente puede ignorar las recompensas futuras y buscar una recompensa inmediata, lo que puede llevarlo a tomar acciones subóptimas. Si el factor de descuento temporal es demasiado alto, el agente puede ser demasiado cauteloso y tardar demasiado en aprender. Por lo tanto, elegir un factor de descuento adecuado es importante para minimizar la trayectoria del agente.
