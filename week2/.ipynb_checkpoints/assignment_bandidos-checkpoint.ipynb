## 1. Bandido greedy ##

### 1.1. Descripcion

La definición de los bandidos consiste en crear un agente, como una clase de Python, `Bandit`. Inicialmente crearemos un bandido con una cantidad de brazos dado por parámetro (por defecto utilizaremos 10 brazos). 
Cada uno de los brazos del bandido (definidos en una lista `arms` como un atributo de la clase) tiene su propia recompensa, la cual asignaremos como un número flotante aleatorio en el rango `[-3,3]` (utilizando la función `random.uniform(-3,3)` de la librería random). Adicionalmente, definimos un atributo de la clase (`reward`) para mantener la recompensa cumulativa del bandido.
Inicialmente nuestro bandido tiene dos funciones `choose_arm` y `run`.

- `choose_arm` no tienen parámetros. Esta función se encarga de seleccionar un brazo a ejecutar (de forma aleatoria en caso de empate), y retorna la recompensa recibida de acuerdo a la fórmula:

![title](img/bandit_formula.png)

Retornando la recompensa promedio obtenida por el agente al escoger el brazo dado.

Para poder realizar este cálculo, debemos llevar la cuenta de la cantidad de veces que ha sido escogido cada brazo y la recompensa acumulada de cada brazo. Esta información la definimos como atributos de la clase, en las variables `occurrences` y `cumulative_rewards`,  respectivamente. Ambos atributos se definen como una lista de tamaño número de brazos del bandido.

El cálculo de la recompensa primero actualiza las ocurrencias de escoger el brazo seleccionado aleatoriamente, `arm`. Además debemos actualizar la recompensa cumulativa de `arm`, sumando la recompensa correspondiente al brazo (obtenida de la lista de brazos). Finalmente, calculamos Q<sub>t</sub>(a) como la recompensa cumulativa sobre las ocurrencias para el brazo dado y retornamos dicho valor. La recompensa global del agente se actualiza con el valor calculado.

- `run` no recibe ningún parámetro. Esta función define la cantidad de episodios para los cuales vamos a calcular la recompensa del agente (1000 en nuestro caso)
Luego creamos un bandido (`bandit`) y una variable para almacenar la recompensa obtenida en cada iteración (`expected_reward`).
Para cada uno de los episodios, ejecutamos el bandido llamando la función choose_arm y guardamos el resultado en la lista de recompensas esperadas.
Finalmente retornamos la lista de recompensas para cada iteración (`expected_reward`).

### 1.2. Analisis de la Implementacion

Para la implementacion de este agente se creo entonces la clase Bandit, que hace uso de la formula mencionada anteriormente. Por la forma en como se implementó, no tiene una etapa específica de exploración, esto se debe a que siempre revisa cual es el brazo que mas recompenza le da, como inicualmente todos tienen un valor de cero, elige el primero, y en caso de que el primero tenga valores negativos, pasa al segundo, y asi sucesivamente, hasta que encuentra un brazo cuya recompensa sea positiva, en este caso, nunca puede asegurar que la recompensa y el brazo que encuentra como el mejor efectivamente lo sea. Dado que la recompensa es estacionaria, en el mejor de los casos todos los brazos tienen recompensa negativa, menos el ultimo que coincide en tener la unica recompensa positiva, y por lo tanto, la mas alta. En el peor de los casos, todas las recompensas son positivas y la primera que encuentra es la menor.

```python
# ponerlo full greedy
class Bandit:
    def __init__(self, arms=10):
        self.arms = np.zeros(shape=arms)
        self.occurrences = np.zeros(shape=arms)
        self.cumulative_rewards = np.zeros(shape=arms)
        for i in range(0, arms):
            self.arms[i]=random.uniform(-3,3)
        self.reward = 0

    def choose_arm(self):
        arm = list(self.cumulative_rewards).index(np.max(self.cumulative_rewards))
        self.occurrences[arm] += 1
        self.cumulative_rewards[arm] += (1/self.occurrences[arm])*(self.arms[arm]- self.cumulative_rewards[arm]) # Qt(a)
        self.reward += self.arms[arm]
        return self.arms[arm] 
            
    def run(self,episodes):
        expected_reward = [0]*episodes
        cum_rewards = [0]*episodes
        for i in range(0, episodes):
            expected_reward[i] = self.choose_arm()
            cum_rewards [i] = self.reward
        return [expected_reward, cum_rewards]
```

```python
b = Bandit()
b.run(10)
```

## 2. Bandidos con &epsilon;-decay ##

### 2.1. Descripcion

Implemente el algoritmos de los bandidos ahora utilizando la estrategia de epsilon decay para mediar entre la exploración y la explotación de los bandidos.
Recuerde que esta nueva estrategia se basa en escoger la acción con mayor recompensa esperada Q<sub>n+1</sub> con probabilidad &epsilon; y una acción aleatoria con probabilidad 1-&epsilon;.

Para ello, modifique el bandido anterior agregando un nuevo atributo `epsilon` con un valor de 0.01 por defecto. Además agregue una nueva función `choose_arm` que retorna la acción a tomar, siguiendo las especificaciones del valor &epsilon;.

### 2.2. Analisis de la Implementacion

Para la implementacion del agente se creo la clase Bandit_Epsilon que a diferencia de la anterior primero tiene un periodo de exploracion que equivale al 10% de los episodios que entran por parametro. Adicionalmente, se hace uso de la funcion `np.random.binomial(1, self.epsilon)` que determina que ese valor sea 1 con una probabilidad de epsilon, este valor al que llamamos explore, es el que determina si se explora un brazo aleatorio, o si como en el anterior se elige siempre el maximo. Para esta implementación, se puede ver, que con un epsilon muy pequeño tiende a tener el mismo comportamiento que la implementación anterior, pues con muy baja probabilidad va a explorar otros brazos. Por lo tanto, para que esta implementación funcione bien, toca buscar de forma eurística el mejor epsilon para que explore lo suficiente, pero no tanto, de forma que encuentre la opción más optima. Esta implementación aun asi mejora con respecto a la anterior.

```python
class Bandit_Epsilon:
    def __init__(self, arms=10):
        self.arms = np.zeros(shape=arms)
        self.occurrences = np.zeros(shape=arms)
        self.cumulative_rewards = np.zeros(shape=arms)
        for i in range(0, arms):
            self.arms[i]=random.uniform(-3,3)
        self.reward = 0
        self.epsilon = 0.01

    def choose_arm(self, exploration_time, episode):
        explore = np.random.binomial(1, self.epsilon)
        if episode >= exploration_time and explore == 0:
            arm = list(self.cumulative_rewards).index(np.max(self.cumulative_rewards))
        else:
            arm = random.randint(0,len(self.arms)-1)
        self.occurrences[arm] += 1
        self.cumulative_rewards[arm] += (1/self.occurrences[arm])*(self.arms[arm]- self.cumulative_rewards[arm])
        self.reward += self.arms[arm]
        return self.arms[arm] # Qt(a)
            
    def run(self,episodes):
        exploration_time= episodes*0.1
        expected_reward = [0]*episodes
        cum_rewards = [0]*episodes
        for i in range(0, episodes):
            expected_reward[i] = self.choose_arm(exploration_time, i)
            cum_rewards [i] = self.reward
        return [expected_reward, cum_rewards]
```

```python
b = Bandit_Epsilon()
b.run(10)
```

## 3. Bandido con recompensas no estacionarias #

### 3.1 Descripción
Ahora retomaremos el bandido del punto 2. para permitir que las recompensas sean no estacionarias.
Para lograr este objetivo, debemos realizar las siguientes modificaciones sobre el bandido.
1. Agregar un atributo con los valores iniciales de recompensa para cada brazo. Con este proposito podemos utilizar la el atributo `arms`, que tenia la recompensa de cada brazo.
Por defecto le daremos un valor inicial a cada brazo, como un entero aleatorio entre 1 y el número de brazos.
2. Debemos introducir un atrobuto &alpha; para el bandido que tendra como valor por defecto `0.1`
3. Dentro de la función `expected_reward`, debemos modificar la recompensa de cada episodio para cada brazo. Para ello utilizaremos una función sinusoidal que nos da la recompensa actual como el seno de la multiplicación del paso actual con el brazo. Todo esto como un factor de la recompensa inicial del brazo.
4. Finalmente, dentro de la función `expected_reward`, debemos realizar el cálculo de la recompensa esperada según la fórmula de las recompensas no estacionarias y almacenarla en el atributo `cumulative_rewards` para cada brazo. 

### 3.2 Analisis de la Implementacion

Para la implementacion del agente se creo la clase Bandit_NoStationary que a diferencia de las demas, las recompensas son variables, y estas varian dependiendo de la siguiente formula `self.arms[arm]*math.sin(arm*self.arms[arm])`, y el valor de las recompensas acumulativas varia segun la siguiente formula:


Adicionalmente, se basa en la implementacion anterior, donde hace una exploracion inicial del 10% de los episodios, y ya despues con una probabilidad de epsilon hace mas exploracion, no obstante, este no tiene mucho problema, pues dado que las recompensas varian, se garantiza que no siempre se va a escoger la misma recompensa como si lo hacen las anteriores implementaciones

```python
class Bandit_NoStationary:
    def __init__(self, arms=10):
        self.arms = np.zeros(shape=arms)
        self.occurrences = np.zeros(shape=arms)
        self.cumulative_rewards = np.zeros(shape=arms)
        for i in range(0, arms):
            self.arms[i]=random.uniform(1,arms)
        self.reward = 0
        self.epsilon = 0.01
        self.alfa = 0.1
    
    def expected_reward(self, arm):
        self.cumulative_rewards[arm] = self.cumulative_rewards[arm]+ self.alfa*(self.arms[arm]-self.cumulative_rewards[arm])
        self.arms[arm] = self.arms[arm]*math.sin(arm*self.arms[arm])

    def choose_arm(self, exploration_time, episode):
        explore = np.random.binomial(1, self.epsilon)
        if episode >= exploration_time and explore == 0:
            arm = list(self.cumulative_rewards).index(np.max(self.cumulative_rewards))
        else:
            arm = random.randint(0,len(self.arms)-1)
        self.occurrences[arm] += 1
        self.expected_reward(arm)
        self.reward += self.arms[arm]
        return self.arms[arm] # Qt(a)
            
    def run(self, episodes):
        exploration_time= episodes*0.1
        expected_reward = [0]*episodes
        cum_rewards = [0]*episodes
        for i in range(0, episodes):
            expected_reward[i] = self.choose_arm(exploration_time, i)
            cum_rewards [i] = self.reward
        return [expected_reward, cum_rewards]
```

```python
b = Bandit_NoStationary()
b.run(10)
```

## 4. Análisis comparativo de las tres implementaciones

A continuacion, se presenta el ambiente de pruebas para cada uno de las implementaciones, se muestra las recompensas de cada uno de los Bandits, y las recompensas acumuladas de cada uno de ellos. De esto, se puede evidenciar que en su mayoria todos los algorimos siempre convergen. Sin embargo, entre mas iteraciones/episodios se pongan, mas empeoran en cuanto a llegar a la recompensa maxima. Una posible hipotesis del porque sucede esto es que los algoritmos se esten overfiteando para llegar a una recompensa especifica. 

Adicionalmente, se puede evidenciar que la mejor implementacion es la del Epsilon_Decay, pues cada uno de los bandidos tiende a converger a la recompensa acumulada maxima, todo esto claramente depende de los valores iniciales con los que comience. Lo que genera un caso de estudio interesante, pues quiere decir que el rendimiento del algoritmo no solo va a depender de la forma como actualice o el brazo que escoja, sino tambien de los puntos iniciales, pues dependiendo de esto puede no converger a la recompensa maxima esperada.

La segunda mejor implementacion es la greedy, que siempre se mantiene en el primer dato positivo que encuentra, y que tiene sentido que Epsilon_Decay sea mejor, pues hacer una mayor exploracion, lo que hace que tenga mas informacion para tomar la decision de cual es el maximo.

Finalmente, la peor implementacion es la de Bandit_NoStationary, esto tiene sentido, pues al ser las recompensas no estacionarias, introduce un mayor grado de aleatoriedad, que no permite que se pueda encontrar la forma más optima para encontrar la mejor recompensa. No obstante, es sorprendente que, aun asi las implementaciones tienden a converger en un punto.


```python
# Episodes
episodes = 100
non_stationary_episodes =100

# Number of bandits
num_bandits = 4
# Numer of arms
num_arms = 100

# x axis data
bandits_x = np.linspace(0, episodes, episodes)
non_stationary_x = np.linspace(0, non_stationary_episodes, non_stationary_episodes)

# plot definition
fig, axs = plt.subplots(2, 3, figsize=(30,12))
# plt.title("Bandit average reward")
plt.xlabel("Iterations")
plt.ylabel("Average reward")

# y axis data definition and graph definition 
def plot_bandit(ax, bx, x, y, title, i):
    ax.plot(x, y[0], linewidth=1.0, label=f'bandit {i}')
    ax.set_title(f'{title} Bandits rewards by iteration')
    ax.legend()
    bx.plot(x, y[1], linewidth=1.0, label=f'bandit {i}')
    bx.set_title(f'{title} Bandits cumulative rewards by iteration')
    bx.legend()
                       
def plot_bandits(ax, bx, tipo, bandits=1):
    for i in range(1, bandits+1):
        if tipo == 1:
            b = Bandit(num_arms)
            y = b.run(episodes)
            plot_bandit(ax, bx, bandits_x, y, "Greedy", i)
        elif tipo == 2:
            b = Bandit_Epsilon(num_arms)
            y = b.run(episodes)
            plot_bandit(ax, bx, bandits_x, y, "Epsilon Decay", i)
        elif tipo == 3:
            b = Bandit_NoStationary(num_arms)
            y = b.run(non_stationary_episodes)
            plot_bandit(ax, bx, non_stationary_x, y, "Non Stationary Decay", i)
        else:
            print("No es un tipo valido")
        # y = np.array(b.run(episodios))
        # ax.plot(bandits_x, y, linewidth=1.0, label=f'bandit {i}')

#example: plot with 3 bandits
#remove the coment for the case you want to observe
# Greedy
plot_bandits(axs[0,0],axs[1,0],1,num_bandits)
# Epsilon Decay
plot_bandits(axs[0,1],axs[1,1],2,num_bandits)
# Non Stationary Decay
plot_bandits(axs[0,2],axs[1,2],3,num_bandits)
                       
# ax.legend()
plt.show()
```
